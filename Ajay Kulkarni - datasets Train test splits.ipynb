## 1. Bank Marketing

#### Abstract: 
The data is related with direct marketing campaigns (phone calls) of a Portuguese banking institution. The classification goal is to predict if the client will subscribe a term deposit (variable y).

#### Data Set Information: 
The data is related with direct marketing campaigns of a Portuguese banking institution. The marketing campaigns were based on phone calls. Often, more than one contact to the same client was required, in order to access if the product (bank term deposit) would be ('yes') or not ('no') subscribed.

#### Attribute Information:
###### Bank client data:

   - Age (numeric)
   - Job : type of job (categorical: 'admin.', 'blue-collar', 'entrepreneur', 'housemaid', 'management', 'retired', 'self-employed', 'services', 'student', 'technician', 'unemployed', 'unknown')
   - Marital : marital status (categorical: 'divorced', 'married', 'single', 'unknown' ; note: 'divorced' means divorced or widowed)
   - Education (categorical: 'basic.4y', 'basic.6y', 'basic.9y', 'high.school', 'illiterate', 'professional.course', 'university.degree', 'unknown')
   - Default: has credit in default? (categorical: 'no', 'yes', 'unknown')
   - Housing: has housing loan? (categorical: 'no', 'yes', 'unknown')
  - Loan: has personal loan? (categorical: 'no', 'yes', 'unknown')

##### Related with the last contact of the current campaign:

   - Contact: contact communication type (categorical: 'cellular','telephone')
   - Month: last contact month of year (categorical: 'jan', 'feb', 'mar', ..., 'nov', 'dec')
  -  Day_of_week: last contact day of the week (categorical: 'mon','tue','wed','thu','fri')
   - Duration: last contact duration, in seconds (numeric). 
Important note: this attribute highly affects the output target (e.g., if duration=0 then y='no'). Yet, the duration is not known before a call is performed. Also, after the end of the call y is obviously known. Thus, this input should only be included for benchmark purposes and should be discarded if the intention is to have a realistic predictive model.

##### Other attributes:

   - Campaign: number of contacts performed during this campaign and for this client (numeric, includes last contact)
   - Pdays: number of days that passed by after the client was last contacted from a previous campaign (numeric; 999 means client was not previously contacted)
   - Previous: number of contacts performed before this campaign and for this client (numeric)
   - Poutcome: outcome of the previous marketing campaign (categorical: 'failure','nonexistent','success')

##### Social and economic context attributes

   - Emp.var.rate: employment variation rate - quarterly indicator (numeric)
   - Cons.price.idx: consumer price index - monthly indicator (numeric)
   - Cons.conf.idx: consumer confidence index - monthly indicator (numeric)
   - Euribor3m: euribor 3 month rate - daily indicator (numeric)
   - Nr.employed: number of employees - quarterly indicator (numeric)

##### Output variable (desired target):

   - y - has the client subscribed a term deposit? (binary: 'yes', 'no')

## 5. Software Defect Prediction

- Number of instances: 10885

- Number of attributes: 22 (5 different lines of code measure, 3 McCabe metrics, 4 base Halstead measures, 8 derived 
          Halstead measures, a branch-count, and 1 goal field)

- Attribute Information:
 
      1. loc             : numeric % McCabe's line count of code
      2. v(g)            : numeric % McCabe "cyclomatic complexity"
      3. ev(g)           : numeric % McCabe "essential complexity"
      4. iv(g)           : numeric % McCabe "design complexity"
      5. n               : numeric % Halstead total operators + operands
      6. v               : numeric % Halstead "volume"
      7. l               : numeric % Halstead "program length"
      8. d               : numeric % Halstead "difficulty"
      9. i               : numeric % Halstead "intelligence"
     10. e               : numeric % Halstead "effort"
     11. b               : numeric % Halstead 
     12. t               : numeric % Halstead's time estimator
     13. lOCode          : numeric % Halstead's line count
     14. lOComment       : numeric % Halstead's count of lines of comments
     15. lOBlank         : numeric % Halstead's count of blank lines
     16. lOCodeAndComment: numeric
     17. uniq_Op         : numeric % unique operators
     18. uniq_Opnd       : numeric % unique operands
     19. total_Op        : numeric % total operators
     20. total_Opnd      : numeric % total operands
     21. branchCount     : numeric % of the flow graph
     22. defects         : {false,true} % module has/has not one or more 
                                        % reported defects

- Missing attributes: none

- Class Distribution: the class value (defects) is discrete 
    false: 2106 = 19.35%
    true:  8779 = 80.65%

## Train-Test Splits

```python
# Bank Marketing Dataset

X = bank_marketing2.iloc[:,0:20]
y = bank_marketing2.y
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=0,stratify = y,test_size = 0.3)

X_train.reset_index(drop=True,inplace=True)
y_train.reset_index(drop=True,inplace=True)
X_test.reset_index(drop=True,inplace=True)
y_test.reset_index(drop=True,inplace=True)

bank_marketing_train = pd.concat([X_train,y_train],axis=1)
bank_marketing_test = X_test
bank_marketing_test_solutions = y_test

print bank_marketing_train.shape
print bank_marketing_test.shape
```

```python
# Car Fuel Cosumption Dataset:

X = car_fuel_cosumption.loc[:,['distance','speed','temp_inside','temp_outside',
       'specials', 'gas_type', 'AC', 'rain', 'sun', 'refill liters',
       'refill gas']]
y = car_fuel_cosumption.consume
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=0,stratify = car_fuel_cosumption.gas_type,test_size = 0.3)

X_train.reset_index(drop=True,inplace=True)
y_train.reset_index(drop=True,inplace=True)
X_test.reset_index(drop=True,inplace=True)
y_test.reset_index(drop=True,inplace=True)

car_fuel_cosumption_train = pd.concat([X_train,y_train],axis=1)
car_fuel_cosumption_test = X_test
car_fuel_cosumption_test_solutions = y_test

print car_fuel_cosumption_train.shape
print car_fuel_cosumption_test.shape
```

```python
# Student Performance Dataset

X = student_performance.loc[:,['school', 'sex', 'age', 'address', 'famsize', 'Pstatus', 'Medu',
       'Fedu', 'Mjob', 'Fjob', 'reason', 'guardian', 'traveltime',
       'studytime', 'failures', 'schoolsup', 'famsup', 'paid',
       'activities', 'nursery', 'higher', 'internet', 'romantic',
       'famrel', 'freetime', 'goout', 'Dalc', 'Walc', 'health',
       'absences', 'G1', 'G2']]
y = student_performance.G3
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=0,stratify = student_performance.sex,test_size = 0.3)

X_train.reset_index(drop=True,inplace=True)
y_train.reset_index(drop=True,inplace=True)
X_test.reset_index(drop=True,inplace=True)
y_test.reset_index(drop=True,inplace=True)

student_performance_train = pd.concat([X_train,y_train],axis=1)
student_performance_test = X_test
student_performance_test_solutions = y_test

print student_performance_train.shape  
print student_performance_test.shape 
```

```python
# Loan Prediction Dataset

X = loan_prediction.loc[:,['Loan_ID', 'Gender', 'Married', 'Dependents', 'Education',
       'Self_Employed', 'ApplicantIncome', 'CoapplicantIncome',
       'LoanAmount', 'Loan_Amount_Term', 'Credit_History', 'Property_Area']]
y = loan_prediction.Loan_Status
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=0,stratify = y,test_size = 0.3)

X_train.reset_index(drop=True,inplace=True)
y_train.reset_index(drop=True,inplace=True)
X_test.reset_index(drop=True,inplace=True)
y_test.reset_index(drop=True,inplace=True)

loan_prediction_train = pd.concat([X_train,y_train],axis=1)
loan_prediction_test = X_test
loan_prediction_test_solutions  = y_test

print loan_prediction_train.shape  
print loan_prediction_test.shape 
```

```python
# Software Defect Prediction Dataset

X = software_defect.loc[:,['loc', 'v(g)', 'ev(g)', 'iv(g)', 'n', 'v', 'l', 'd', 'i', 'e',
       'b', 't', 'lOCode', 'lOComment', 'lOBlank', 'locCodeAndComment',
       'uniq_Op', 'uniq_Opnd', 'total_Op', 'total_Opnd', 'branchCount']]
y = software_defect.defects
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=0,stratify = y,test_size = 0.3)

X_train.reset_index(drop=True,inplace=True)
y_train.reset_index(drop=True,inplace=True)
X_test.reset_index(drop=True,inplace=True)
y_test.reset_index(drop=True,inplace=True)

software_defect_train = pd.concat([X_train,y_train],axis=1)
software_defect_test = X_test
software_defect_test_solutions = y_test

print software_defect_train.shape  
print software_defect_test.shape
```
